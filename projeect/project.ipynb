## Introduction to Sorting Algorithms

Sorting is a fundamental concept that involves arranging elements in a specific order. Sorting algorithms play a crucial role in various applications and understanding their complexity, performance, and characteristics is essential.

Key concepts related to sorting algorithms include:
- Time and space complexity: Sorting algorithms can have different time and space complexities, impacting their efficiency.
- In-place sorting: Some sorting algorithms can sort the elements within the given array without requiring additional space.
- Stable sorting: Stable sorting algorithms maintain the relative order of equal elements in the sorted output.
- Comparator functions: Sorting algorithms may use comparator functions to define custom comparison logic.
- Comparison-based and non-comparison-based sorts: Sorting algorithms can be categorized based on whether they rely on element comparisons for sorting.

Studying these algorithms and their implementations can help us to optimize our approaches to managing datasets, reviewing data, and developing our code.

### Bubble Sort

```python
def bubbleSort(arr):
    n = len(arr) # Get length of the array
    # Iterate through the elements
    for i in range(n-1): # range(n) also works but outer loop repeats more than needed
        swapped = False
        for j in range(0, n-i-1): # Iterate through from 0 to n-i-1
            if arr[j] > arr[j+1]: # swap if the elemebt found is greater than the next 
                swapped = True
                arr[j], arr[j+1] = arr[j+1], arr[j]
        if not swapped: # if no swap required just exit main loop
            return
```

Inspired by the code on [GeeksForGeeks](https://www.geeksforgeeks.org/python-program-for-bubble-sort/).

Bubble Sort is a simple sorting algorithm that repeatedly goes through the list, compares adjacent elements and swaps them if they are in the wrong order. This process is repeated until the list is sorted. The algorithm gets its name because the smaller items will ‘bubble’ to the top of the list (from [Stack Overflow](https://stackoverflow.com/questions/31059479/why-bubble-sort-is-called-bubble-sort#:~:text=The%20bubble%20sort%20gets%20its,bubbles%20rising%20to%20the%20surface.&text=It's%20called%20bubble%20sort%20because,end%2Fbeginning%20of%20an%20array.)). Despite its simplicity and ease of implementation, bubble sort is not efficient for large datasets due to its low time complexity.

Complexity:
- The time complexity of bubble sort is O(n^2) in the worst and average cases (animeshdey, 2024), where n is the number of elements in the list. This is because, in the worst case, each element must be compared with all other elements. The best time complexity is O(n) (animeshdey, 2024), which occurs when the array is already sorted, but even then bubble sort is generally outperformed by other algorithms such as quicksort or mergesort.

- The space complexity of bubble sort is O(1) (animeshdey, 2024), which means that it requires a constant amount of additional space. This is because bubble sort is an in-place sorting algorithm, which sorts the list without requiring additional space for another array.

Despite its inefficiency, bubble sort can be useful for educational purposes to understand the concept of sorting and the mechanics of comparison-based algorithms. However, for practical purposes, more advanced algorithms are recommended for sorting large data sets (Alake, 2023).

It has "horrible" rating on the complexity curve below (image taken from Lecture notes Week 5 pdf).

```python

def bubbleSortwithDiagram(arr):
    n = len(arr)  # Get the length of the array
    # Iterate through the elements
    for i in range(n-1):
        swapped = False
        for j in range(0, n-i-1):  # Iterate through from 0 to n-i-1
            if arr[j] > arr[j+1]:  # Swap if the element found is greater than the next
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        # Visualize the array at the end of each pass
        plt.clf() 
        plt.bar(range(len(arr)), arr, color='blue')
        plt.title(f"Bubble Sort Visualization: Pass {i+1}")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.pause(1)  # Pause to create the visualization effect
        if not swapped:
            return  # Stop if the array is already sorted
    plt.show()
```

### Selection Sort

```python
def selectionSort(arr):
    n = len(arr) 
    for i in range(n): 
        min_idx = i   # find the index of the minimum element (unsorted)
        for j in range(i+1, n):  
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i] # Swap the found minimum element with the first element
```

Inspired by the code found in the C and Python3 subsections on [GeeksforGeeks](https://www.geeksforgeeks.org/selection-sort/)

Selection Sort is a simple sorting algorithm that divides the input list into two parts: the sublist of already sorted items, which is constructed from left to right, and the sublist of items remaining to be sorted, which occupies the remainder of the list (A S, 2023). The algorithm repeatedly selects the smallest (or largest, depending on the sort order) element from the unsorted sublist and moves it to the end of the sorted sublist.

Key Stages in Selection Sort:
- Find the Minimum Element: Traverse the unsorted part of the array to find the minimum element.
- Swap: Swap the minimum element found with the first element of the unsorted part.
- Repeat: Move the boundary between the sorted and unsorted parts of the array one step to the right and repeat until the entire array is sorted.

Complexity:
- The worst-case, best-case and average case of the Selection Sort algorithm have a time complexity of O(n^2)(A S, 2023), where n is the number of elements in the array. This is because there are two nested loops: one that goes through each element and one that looks for the minimum in the unsorted part.

- The space complexity of Selection Sort is O(1), as it only requires a constant amount of additional space for the time variables used for the exchanges(A S, 2023).


```python

def selectionSortDiagrams(arr):
    n = len(arr)
    for i in range(n): 
        min_idx = i # find the index of the minimum element (unsorted)
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Swap the found minimum element with the first element
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        # Visualize the array at the end of each pass
        plt.clf()
        plt.bar(range(len(arr)), arr, color='blue')
        plt.title(f"Selection Sort Visualization: Pass {i+1}")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.pause(1)  # Pause to create the visualization effect
    plt.show()
```

### Insertion Sort

```python
def insertionSort(arr):
    n = len(arr)  # Get the length of the array
    for i in range(1, n):
        key = arr[i]  # Element to be positioned
        j = i - 1
        # Move elements of arr[0..i-1] that are greater than key to one position ahead
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key  # Insert the key element in the correct position
```

This code was inspired by the Python implementation found [here](https://developer.nvidia.com/blog/insertion-sort-explained-a-data-scientists-algorithm-guide/#:~:text=The%20worst%2Dcase%20(and%20average,O(n)%20time%20complexity.)) (Alake, 2021)

Insertion Sort is a simple sorting algorithm that builds a sorted list one by one, moving the unsorted elements to their proper place in the sorted list (w3schools, n.d.). It starts with a single-item list (thinking of it as already sorted) and then inserts a new item from the unsorted list into the correct position in the sorted list.

Key Stages in Insertion Sort:
- Item Selection: Select an item from the unsorted list.
- Compare and Insert: Compares the selected item with the items in the sorted list and inserts it into the correct position in the sorted list.
- Repeat: Repeats the above steps until all items in the unsorted list have been inserted into the sorted list.

Complexity:
- The time complexity of Insertion Sort varies according to the order of the items in the input list. In the best case, when the list is almost ordered, the time complexity is O(n), where n is the number of elements in the list (Alake, 2021). In the worst case, when the list is in reverse order, the time complexity is O(n^2) (Alake, 2021).

- The space complexity of the Insertion Sort is O(1), as it only requires a constant amount of additional space for the time variables used for the exchanges (Brilliant.org, n.d.).

```python

def insertionSortDiagram(arr):
    n = len(arr)  # Get the length of the array
    for i in range(1, n):
        key = arr[i]  # Element to be positioned
        j = i - 1
        # Move elements of arr[0..i-1] that are greater than key to one position ahead
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key  # Insert the key element in the correct position
        
        # Visualize the array at the end of each pass
        plt.clf()
        plt.bar(range(len(arr)), arr, color='blue')
        plt.title(f"Insertion Sort Visualization: Pass {i}")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.pause(1)  # Pause to create the visualization effect
    plt.show()
```

### Merge Sort

```python
def mergeSort(arr):
    # Base case: If the array has one or no elements, it is already sorted
    if len(arr) > 1:
        # Find the middle point and split the array into left and right halves
        mid = len(arr) // 2
        L = arr[:mid]  # Left half
        R = arr[mid:]  # Right half

        # Recursively sort both halves
        mergeSort(L)  # Sort the left half
        mergeSort(R)  # Sort the right half

        # Initialize pointers for L, R, and the main array
        i = j = k = 0

        # Merge the sorted halves back into the main array
        while i < len(L) and j < len(R):
            if L[i] < R[j]:  # If the current element of L is smaller than the current element of R
                arr[k] = L[i]  # Place the smaller element in the main array
                i += 1  # Move the pointer in L forward
            else:
                arr[k] = R[j]  # Place the smaller element in the main array
                j += 1  # Move the pointer in R forward
            k += 1  # Move the pointer in the main array forward

        # If there are any remaining elements in L, copy them to the main array
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        # If there are any remaining elements in R, copy them to the main array
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
```

Adapted from the Python program found [here](https://developer.nvidia.com/blog/merge-sort-explained-a-data-scientists-algorithm-guide/). 

Merge Sort is one of the divide-and-conquer algorithms that sorts an array by dividing it into two halves, sorting each half, and then merging the sorted halves back together (Alake, 2022). This algorithm is efficient for large datasets and has a predictable time complexity.

Key Stages in Merge Sort:

- Divide: Split the array into two halves.
- Conquer: Recursively sort each half.
- Combine: Merge the two sorted halves to produce the sorted array.

Complexity:

- The time complexity of Merge Sort is O(nlogn) for all cases (best, average, and worst) (Harel and Feldman, 2004, p.262), where n is the number of elements in the array. This is because the array is repeatedly divided in half (logarithmic number of divisions) and each level of division requires linear time to merge the elements (Alake, 2022).
- The space complexity of Merge Sort is O(n) (GeeksforGeeks, 2018), as it requires additional space for the temporary arrays used during the merge process.

```python

def mergeSortDiagram(arr):
    """
    Sorts the given array using the Merge Sort algorithm and visualizes each step.
    
    Args:
        arr (list): The array to be sorted.
    """
    # Base case: If the array has one or no elements, it is already sorted
    if len(arr) > 1:
        # Find the middle point and split the array into left and right halves
        mid = len(arr) // 2
        L = arr[:mid]  # Left half
        R = arr[mid:]  # Right half

        # Visualize the left and right halves
        plt.clf()
        plt.subplot(1, 2, 1)
        plt.bar(range(len(L)), L, color='blue')
        plt.title("Left Half")
        plt.xlabel("")
        plt.ylabel("Value")

        plt.subplot(1, 2, 2)
        plt.bar(range(len(R)), R, color='green')
        plt.title("Right Half")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.pause(1)  # Pause to create the visualization effect

        # Recursively sort both halves
        mergeSortDiagram(L)  # Sort the left half
        mergeSortDiagram(R)  # Sort the right half

        # Merge the sorted halves back into the main array
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Copy the remaining elements of L, if any
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        # Copy the remaining elements of R, if any
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

        # Visualization: Plot the merged array
        plt.clf()
        plt.bar(range(len(arr)), arr, color='purple')
        plt.title("Merge Sort Visualization: Merged Array")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.pause(1)  # Pause to create the visualization effect
    plt.show()
```

```python

# Generate a random array based on the student number
array = generate_random_array(10)  # Adjust the size as needed

print("Unsorted array:", array)

# Perform merge sort with visualization
mergeSortDiagram(array)

print("Sorted array:", array)
```

### Counting Sort

```python
def countingSort(arr):
    # Find the maximum value in the array to determine the size of the count array
    max_val = max(arr)

    # Initialize the count array with zeros, with size of max_val + 1
    count = [0] * (max_val + 1)

    # Initialize the output array with zeros, with the same size as the input array
    output = [0] * len(arr)

    # Step 1: Count the occurrences of each element in the input array
    for num in arr:
        count[num] += 1

    # Step 2: Modify the count array by adding the previous counts (cumulative count)
    for i in range(1, max_val + 1):
        count[i] += count[i - 1]

    # Step 3: Build the output array by placing elements in their correct positions
    i = len(arr) - 1
    while i >= 0:
        output[count[arr[i]] - 1] = arr[i]  # Place the element at the correct position in output array
        count[arr[i]] -= 1  # Decrease the count for the placed element
        i -= 1  # Move to the previous element in the input array

    # Step 4: Copy the sorted elements from the output array back to the input array
    for i in range(len(arr)):
        arr[i] = output[i]
```

Inspired and adapted from [GeeksforGeeks](https://www.geeksforgeeks.org/counting-sort/).

Counting Sort is a non-comparative sorting algorithm that sorts the elements of an array by counting occurrances of each unique element. The algorithm assumes that the input elements are non-negative integers (GeeksforGeeks, 2024).

Key Stages in Counting Sort:

- Counting Occurrences: Count the occurrences of each element in the input array.
- Cumulative Count: Modify the count array by computing the cumulative count.
- Building Output Array: Place each element in its correct position in the output array based on the cumulative count.
- Copy to Original Array: Copy the sorted elements from the output array back to the input array.

Complexity:
- The time complexity of Counting Sort is O(n + k), where n is the number of elements in input array and k is the range of the input (i.e., the difference between the maximum and minimum values plus one) . In the best, average, and worst cases, Counting Sort exhibits linear time complexity due to its linear traversal of the input array (Simplilearn, 2023) and the count array.

- The space complexity of Counting Sort is O(n + k), where n is the number of elements in the input array and k is the range of the input (StudySmarter UK, n.d.). Counting Sort requires additional space for the count array and the output array, which together take up space proportional to the size of the input array and the range of the input. However, as the range of the input is usually much smaller than the number of elements in the array (k << n), the space complexity is often considered linear with respect to the size of the input array (GeeksforGeeks, 2024).

```python

def countingSortDiagram(arr):
    """
    Sorts the given array using the Counting Sort algorithm and visualizes each step.
    
    Args:
        arr (list): The array to be sorted.
    """
    # Find the maximum value in the array to determine the size of the count array
    max_val = max(arr)

    # Initialize the count array with zeros, with size of max_val + 1
    count = [0] * (max_val + 1)

    # Initialize the output array with zeros, with the same size as the input array
    output = [0] * len(arr)

    # Visualization: Plot the initial array
    plt.clf()
    plt.bar(range(len(arr)), arr, color='blue')
    plt.title("Counting Sort Visualization: Initial Array")
    plt.xlabel("")
    plt.ylabel("Value")
    plt.pause(1)  # Pause to create the visualization effect

    # Step 1: Count the occurrences of each element in the input array
    for num in arr:
        count[num] += 1

    # Visualization: Plot the count array
    plt.clf()
    plt.bar(range(len(count)), count, color='green')
    plt.title("Counting Sort Visualization: Count Array")
    plt.xlabel("")
    plt.ylabel("Count")
    plt.pause(1)  # Pause to create the visualization effect

    # Step 2: Modify the count array by adding the previous counts (cumulative count)
    for i in range(1, max_val + 1):
        count[i] += count[i - 1]

    # Visualization: Plot the modified count array
    plt.clf()
    plt.bar(range(len(count)), count, color='orange')
    plt.title("Counting Sort Visualization: Modified Count Array")
    plt.xlabel("")
    plt.ylabel("Cumulative Count")
    plt.pause(1)  # Pause to create the visualization effect

    # Step 3: Build the output array by placing elements in their correct positions
    i = len(arr) - 1
    while i >= 0:
        output[count[arr[i]] - 1] = arr[i]  # Place the element at the correct position in output array
        count[arr[i]] -= 1  # Decrease the count for the placed element
        i -= 1  # Move to the previous element in the input array

        # Visualization: Plot the array after each element is placed
        plt.clf()
        plt.bar(range(len(output)), output, color='red')
        plt.title("Counting Sort Visualization: Sorted Array")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.pause(0.1)  # Pause to create the visualization effect

    # Copy the sorted elements from the output array back to the input array
    for i in range(len(arr)):
        arr[i] = output[i]

    # Visualization: Plot the final sorted array
    plt.clf()
    plt.bar(range(len(arr)), arr, color='purple')
    plt.title("Counting Sort Visualization: Final Sorted Array")
    plt.xlabel("")
    plt.ylabel("Value")
    plt.pause(1)  # Pause to create the visualization effect
plt.show()
```

```python

# Generate a random array based on the student number
array = generate_random_array(10)  # Adjust the size as needed

print("Unsorted array:", array)

# Perform counting sort with visualization
countingSortDiagram(array)

print("Sorted array:", array)
```

## Benchmarking

```python
def benchmark_display():
    # Define the sizes of input arrays to test
    sizes = [10, 100, 250, 500, 1000, 1250, 2000, 3000, 5000, 6000, 8000]
    # List of sorting algorithms to benchmark
    algorithms = [bubbleSort, selectionSort, insertionSort, mergeSort, countingSort]
    # Retrieve the names of the algorithms for labeling purposes
    algorithm_names = [algorithm.__name__ for algorithm in algorithms]
    # Dictionary to store the average time taken by each algorithm for each size
    time_taken = {algorithm_name: [] for algorithm_name in algorithm_names}
    # List to collect all results for displaying in a DataFrame
    results = []

    # Loop through each size
    for size in sizes:
        row = {'Input Size': size}  # Initialize a row for the current size
        # Test each algorithm with the current size
        for algorithm in algorithms:
            total_time = 0  # Initialize total time for the current algorithm and size
            # Repeat the sorting process 10 times to average the results
            for _ in range(10):
                arr = generate_random_array(size)  # Generate a random array of the current size
                start_time = time.time()  # Record the start time
                algorithm(arr)  # Sort the array using the current algorithm
                end_time = time.time()  # Record the end time
                total_time += (end_time - start_time) * 1000  # Calculate the time taken and convert to milliseconds
            average_time = total_time / 10  # Calculate the average time
            row[algorithm.__name__] = f"{average_time:.2f} ms"  # Store the average time in the row
            time_taken[algorithm.__name__].append(average_time)  # Append the average time to the time_taken dictionary
        results.append(row)  # Append the row to the results list

    df = pd.DataFrame(results)  # Create a DataFrame from the results
    display(df)  # Display the DataFrame

    # Plot the results for each algorithm
    for algorithm_name in algorithm_names:
        plt.plot(sizes, time_taken[algorithm_name], label=algorithm_name, marker='o')  # Plot the average times for each size

    plt.xlabel('Input Size (n)')  # Label the x-axis
    plt.ylabel('Average Time (ms)')  # Label the y-axis
    plt.title('Sorting Algorithms Benchmark')  # Title of the plot
    plt.legend(title = "Algorithm", loc="upper left")  # Add a legend with a title
    plt.grid(True)  # Enable grid for better readability
    plt.show()  # Display the plot
    
```

```python
benchmark_display()
```
